### Determining missing data presence and Nature

```python
#look for missing data
df.isnull().sum()
```

```python
msno.bar(df) 
```

The graph shows that bed, bath, acre_lot and house_size have 20% missing data on average while prev_sold_date has approximately 60% missing. The most likely explanation is that only about 40% of the houses have been sold so far, we can consider the missing data "missing at random (MAR)".

In order to determine the nature of missing data of the other mentioned variables, I will do some visualization

```python
msno.heatmap(df)
```

Making a heatmap of the missing variables, we find correlations between the missing data of each variables. According to msno documentation, a value of <1 means that two variables have a correlation close to 1 but not quite. In other words, If bed has data, it is likely bath will have missing data and vice versa. Bed and bath also have <1 when it comes to house size. From this information, we can deduce that the missing data is most likely MAR but because the correlation is close to 1 but not exactly 1, there might be other factors at play which causes the missing data.

Logically speaking, house sizes that are smaller could cause it to not have a bedroom or a bathroom and instead, have a singular multi-purpose room. Assuming data is MAR, I will use MICE in order to fill in the missing data. In reference to the correlation matrix, I will not be using "status", "city" and "states" in the imputation as they do not seem to affect variables with missing data in any meaningful way.

```python
#store the dropped columns for later
status = df['status']
prev_sold_date = df['prev_sold_date']
city = df['city']
state = df['state']

#impute the missing data from remaining columns
df_nocat = df.drop(['status','prev_sold_date', 'city', 'state'], axis=1)
df_imputed = MICE().fit_transform(df_nocat)
```

```python
# Convert the result back to DataFrame
df_imputed = pd.DataFrame(df_imputed, columns = df_nocat.columns)

# Add the dropped columns back to the DataFrame
df_imputed['status'] = status.values
df_imputed['prev_sold_date'] = prev_sold_date.values
df_imputed['city'] = city.values
df_imputed['state'] = state.values
```

```python
print(df_imputed.isnull().sum())
```

### Outlier treatment

```python
# apply winsorizing to each column
columns = ['bed', 'bath', 'acre_lot', 'house_size','price']
for col in columns:
    df_imputed[col] = winsorize(df_imputed[col], limits=[0.01, 0.01])
```

After winsorizing our numeric variables, we are able to replace potential outliers with values at the 1%. Even these values are less than the above mentioned property but for the purposes of this analysis, it is sufficient.

### How can we predict price?

For this analysis, I elect to perform a regression using XGBoost and random forest

```python
df_imputed.head()
```

#### Feature Engineering

In order to capture as much of the effect of variables on price per square feet, I will be introducing some feature engineering where I make interaction terms, polynomial variables and encoding for efficient data processing.

```python
#Sold column
df_final = df_imputed.drop(['zip_code'], axis=1)

#Define a model that checks whether a property is sold or not
def sold_state(d): 
    if pd.notnull(d):
        return 'yes'
    else:
        return 'no'

df_final['sold'] = df_final['prev_sold_date'].apply(sold_state) 
```

Above is a new variable "sold" which represents whether a property being sold or not. This is is to capture the effect of a proprty being sold has on price per square feet without having to use the date variable.

```python
#polynomial variables
df_final['bed2'] = df_final['bed']**2
df_final['bath2'] = df_final['bath']**2
df_final['house_size2'] = df_final['house_size']**2
```

In order to capture non-linear effects of the number of beds and bathrooms, I have created polynomial variables bed2, bath2 and house_size2 which are the number of rooms or sqfeet squared. In theory, an additional bed or bathroom might not increase price in the same way.

```python
#label encoding for city
le = LabelEncoder()
df_final['city_enc'] = le.fit_transform(df_final['city'])

#label mapping saved
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))

#create dictionary where keys are numeric label and values are cities
reverse_label_mapping = dict(zip(label_mapping.values(), label_mapping.keys()))
```

```python
#Check city value
city_name = reverse_label_mapping[3]
city_name
```

City is label-encoded where each city is assigned a number, this is to make cities into a numerical variable without having to do one-hot encoding which introduces far too many variables. A new df is created which stores all city names with their respective numbers.

```python
df_final = pd.get_dummies(df_final, columns = ['status', 'state', 'sold'], drop_first=True)
```

```python
df_final.head()
```

```python
#splitting the data
X, y = df_final.drop(columns = ['prev_sold_date','city','price','price/sqfeet']), df_final['price/sqfeet']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# Parameter dictionary specifying base learner
param_grid = {
    'n_estimators': [50, 100, 150],
}

model = XGBRegressor()
grid_search = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=5, verbose=1, n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python
best_model = grid_search.best_estimator_
predictions = best_model.predict(X_test)
mse = MSE(y_test, predictions)
print("Mean Squared Error:", mse)
```
